```python
import pandas as pd
```

```python
df = pd.read_csv(r'raw_employee_attrition.csv')
```

```python
df.columns = [col.lower() for col in df.columns]
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('-', '_')
```

```python
for col in df.select_dtypes(include=['object', 'category']).columns:
    print(f"\n=== {col} ===")
    print(df[col].value_counts(dropna=False))
```

```python
# There's something wrong with 'company_tenure'.
# Also, there's something weird with 'years_at_company'.. in some cases..

df_cleaned = df.drop(columns = ['years_at_company', 'company_tenure'], axis = 1)
```

```python
# Transform boolean columns into numerical

df_cleaned['gender'] = df_cleaned['gender'].replace({'Male': 1, 'Female': 0}).astype(int)
df_cleaned['overtime'] = df_cleaned['overtime'].replace({'Yes': 1, 'No': 0}).astype(int)
df_cleaned['remote_work'] = df_cleaned['remote_work'].replace({'Yes': 1, 'No': 0}).astype(int)
df_cleaned['leadership_opportunities'] = df_cleaned['leadership_opportunities'].replace({'Yes': 1, 'No': 0}).astype(int)
df_cleaned['innovation_opportunities'] = df_cleaned['innovation_opportunities'].replace({'Yes': 1, 'No': 0}).astype(int)

df_cleaned['attrition'] = df_cleaned['attrition'].replace({'Left': 1, 'Stayed': 0}).astype(int)
```

```python
for col in df_cleaned.select_dtypes(include='object').columns:
    df_cleaned[col] = df_cleaned[col].astype('category')
```
