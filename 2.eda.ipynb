```python
cat_cols = ['gender', 'job_role', 'work_life_balance', 'job_satisfaction', 'performance_rating', 'number_of_promotions', 'overtime', 'education_level', 
            'marital_status', 'number_of_dependents', 'job_level', 'company_size', 'remote_work', 'leadership_opportunities', 'innovation_opportunities', 
            'company_reputation', 'employee_recognition', 'attrition']

for col in cat_cols:
    plt.figure(figsize = (10,4))
    sns.countplot(x=col, data = df, order = df[col].value_counts().index)
    plt.xticks(rotation = 45)
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
#Resignation rates by categorical features
```

```python
cat_cols = ['gender', 'job_role', 'work_life_balance', 'job_satisfaction', 'performance_rating', 'number_of_promotions', 'overtime', 'education_level', 
            'marital_status', 'number_of_dependents', 'job_level', 'company_size', 'remote_work', 'leadership_opportunities', 'innovation_opportunities', 
            'company_reputation', 'employee_recognition', 'attrition']

for col in cat_cols:
    resignation_rate = df.groupby(col)['attrition'].mean().sort_values(ascending=False)
    resignation_rate.plot(kind='bar', title=f'Resignation Rate by {col}', figsize=(10, 5))
    plt.ylabel('Resignation Rate')
    plt.show()
```

Correlation heatmap

```python
df.nunique()
```

```python
numeric_cols = [
    'age', 'monthly_income', 'number_of_promotions', 'overtime',
    'distance_from_home', 'number_of_dependents', 'remote_work',
    'leadership_opportunities', 'innovation_opportunities', 'attrition'
]

corr = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation Heatmap (Numerical + Binary Features)')
plt.show()
```

```python
# Most Insightful Graphs for Attrition Prediction:
# Resignation Rate by job_level: Shows a very strong trend: Entry-level employees have significantly higher attrition than Mid or Senior levels. This is a critical predictor.

# Resignation Rate by age: Clearly demonstrates that younger employees (19-25) have the highest resignation rates, which steadily decrease with age. Another strong, actionable insight.

# Resignation Rate by marital_status: Reveals that Single employees have a substantially higher resignation rate than Divorced or Married individuals. This is a powerful differentiator.

# Resignation Rate by work_life_balance: Indicates that "Poor" and "Fair" work-life balance strongly correlate with higher attrition, while "Good" and "Excellent" significantly reduce it. Very actionable for policy changes.

# Resignation Rate by job_satisfaction: Highlights that "Low" job satisfaction leads to high attrition. Also, the "Very High" category warrants a closer look as it also shows a high rate.

# Resignation Rate by performance_rating: Shows that "Low" and "Below Average" performance ratings are linked to higher attrition, an expected but important finding.

# Resignation Rate by overtime: A clear signal that working overtime significantly increases the resignation rate. Crucial for understanding employee well-being and retention.

# Resignation Rate by number_of_promotions: Demonstrates that 0, 1, or 2 promotions are associated with high attrition, while 3+ promotions drastically reduce it. This is a key indicator of career progression's impact.

# Resignation Rate by education_level: Distinctly shows that employees with a PhD have a much lower resignation rate compared to all other education levels, which have similar high rates.

# Resignation Rate by company_reputation: Indicates that a "Poor" or "Fair" company reputation is associated with higher attrition than "Good" or "Excellent."

# Resignation Rate by number_of_dependents: Reveals that employees with fewer dependents (0-3) tend to have higher attrition than those with more (4-6).
```
